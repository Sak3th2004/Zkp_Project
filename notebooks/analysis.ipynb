# Metrics Analysis
This notebook summarizes proof and verification performance collected from the SQLite metrics table for advanced zero knowledge simulations.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

db_path = Path("../zkp.db").resolve()
if not db_path.exists():
    raise FileNotFoundError("Run the backend simulations to populate metrics")
with sqlite3.connect(db_path) as conn:
    df = pd.read_sql_query("SELECT app_type, input_size, proof_time, verify_time FROM metrics", conn)
summary = df.groupby("app_type").mean(numeric_only=True)
summary
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
for app, group in df.groupby("app_type"):
    group = group.sort_values("input_size")
    ax.plot(group["input_size"], group["proof_time"], label=f"{app} proof")
    ax.plot(group["input_size"], group["verify_time"], linestyle="--", label=f"{app} verify")
ax.set_xlabel("Input Size")
ax.set_ylabel("Time (s)")
ax.legend()
fig
```
